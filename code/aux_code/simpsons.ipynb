```python
# Your target variables and input features
target_variables = ['CH3Br', 'CO', 'Ethane', 'O3', 'CH4', 'DMS']
input_features = [
    'mean_distance_to_cluster_centroids', 'mean_distance_to_antimeridian', 'mean_distance_to_t1000_cities',
    'min_distance_to_t1000_cities', 'min_distance_to_cluster_centroids', 'min_distance_to_extent_corners',
    'Specific_Humidity_0', 'Potential_Temperature_0', 'mean_lcl', 'bearings_from_origin_-24',
    'sum_moisture_flux', 'mean_mixing_depth', 'sum_solar_radiation', 'domain_indicator', 'time_difference_seconds_0'
]

# Output directory
output_dir = r"C:\Users\vwgei\Documents\PVOCAL\plots\simpsons"
os.makedirs(output_dir, exist_ok=True)  # Ensure the directory exists

# Unique values of spatial_cluster
spatial_clusters = df['spatial_cluster'].unique()

# Loop through spatial clusters, target variables, and input features
for cluster in spatial_clusters:
    cluster_data = df[df['spatial_cluster'] == cluster]  # Filter data for the current spatial cluster
    
    for target in target_variables:
        for feature in input_features:
            # Create the scatter plot
            plt.figure(figsize=(8, 6))
            plt.scatter(cluster_data[feature], cluster_data[target], alpha=0.7, edgecolor='k')
            plt.title(f'Spatial Cluster: {cluster}, {feature} vs {target}')
            plt.xlabel(feature)
            plt.ylabel(target)
            plt.grid(True)
            
            # Save the plot with a unique name
            file_name = f"cluster_{cluster}_{feature}_vs_{target}.png"
            file_path = os.path.join(output_dir, file_name)
            plt.savefig(file_path)
            plt.close()  # Close the plot to free memory
```
