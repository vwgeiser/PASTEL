```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
# Filter out any rows with missing coordinates or Ethane values for clean plotting
plot_df = combined_df.dropna(subset=['LAT', 'LON', 'Ethane_WAS'])

# Set up a geographic plot using Cartopy
fig, ax = plt.subplots(figsize=(12, 8), subplot_kw={'projection': ccrs.PlateCarree()})

# Add natural earth features for context (like coastlines and countries)
ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linestyle=':', linewidth=0.5)
ax.add_feature(cfeature.LAND, edgecolor='black', facecolor='lightgray', alpha=0.3)
ax.add_feature(cfeature.OCEAN, edgecolor='black', facecolor='lightblue', alpha=0.1)

# Scatter plot with color representing 'Ethane_WAS' concentration
scatter = ax.scatter(
    plot_df['LON'], 
    plot_df['LAT'], 
    c=plot_df['Ethane_WAS'], 
    cmap='viridis', 
    s=5, 
    alpha=0.7,
    transform=ccrs.PlateCarree()
)

# Add color bar
cbar = plt.colorbar(scatter, ax=ax, orientation='vertical', shrink=0.7, pad=0.05)
cbar.set_label('Ethane (pptv)')

# Set title and labels
ax.set_title("Georeferenced Scatter Plot of Ethane Concentration", fontsize=14)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.grid(True)
# Show the plot
plt.show()
```
